```python
import  numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import pandas as pd
import random


#算损失函数
def J(theta,X_b,y):

    print(np.sum((y-X_b.dot(theta))**2)/len(X_b))
    return (np.sum((y-X_b.dot(theta))**2))/len(X_b)



#求导
def DJ(theta,X_b,y):
    statis=np.empty(len(theta))

    statis[0]= np.sum((X_b.dot(theta)-y))


    for i in range(1,len(theta)):
        statis[i]=((X_b.dot(theta)-y)).dot(X_b[:,i])
        
    return 2*statis/len(X_b)

#梯度下降

def gradient_descent(theta,learn_rate,X_b,y,limit=1e-8,times=1e4):

    times_sic=0

    while times_sic<times:

          thetaFilly = theta
          theta=theta-learn_rate*DJ(theta,X_b,y)
          # print(abs(J(theta,X_b,y)-J(thetaFilly,X_b,y)))
          if abs(J(theta,X_b,y)-J(thetaFilly,X_b,y))<limit:
              break;


          times_sic+=1

    return theta
```
